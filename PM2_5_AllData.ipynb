```python
from datetime import datetime

custom_date_parser = lambda x: datetime.strptime(x, "%y%m%d")

df3 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data Science/all_data.csv',
                    parse_dates=['วัน/เดือน/ปี'],
                    date_parser=custom_date_parser)
df3
```

```python
df3.info()
```

```python
df3['ชั่วโมง'] = df3['ชั่วโมง'].astype(str)
```

```python
df3['ชั่วโมง'] = df3['ชั่วโมง'].replace(['2400', '100','200'],['0000', '0100', '0200'])
```

```python
df3[0:20]
```

```python
tt=[]

for i in range(len(df3)):
  t = [df3['ชั่วโมง'][i]]
  x = pd.to_datetime(t, format="%H%M")
  y = x.strftime("%H:%M").tolist()
  tt.append(y)
```

```python
df3['time'] = pd.DataFrame(tt)
```

```python
df3
```

```python
df3['Timestamp'] =df3['วัน/เดือน/ปี'].astype(str) + ' ' + df3['time']
df3.head()
```

```python
df3['Timestamp'] = pd.to_datetime(df3['Timestamp'])
```

```python
df3.info()
```

```python
#จัดรูปให้ดูง่ายขึ้น

df3['Year'] = [year.strftime("%Y") for year in df3['Timestamp']]
df3['Month'] = [month.strftime("%Y-%m") for month in df3['Timestamp']]
df3['Date'] = [date.strftime("%Y-%m-%d") for date in df3['Timestamp']]
df3['Time'] = [time.strftime("%H:%M") for time in df3['Timestamp']]

df3.drop(['วัน/เดือน/ปี', 'ชั่วโมง', 'time'], axis=1, inplace=True)

df3[0:10]
```

```python
df3.tail()
```

```python
df3.drop(df3[df3['Year'] == '2021'].index, inplace=True)
df3.tail()
```
